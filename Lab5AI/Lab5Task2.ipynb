```python
import pandas as pd
import numpy as np
from sklearn.linear_model import LinearRegression
import matplotlib.pyplot as plt

data = {
    'Area': [1500, 1800, 2400, 3000, 3500, 4000, 4500, 5000, 6000, 7000],
    'Price': [400000, 450000, 500000, 550000, 600000, 650000, 700000, 750000, 800000, 850000]
}

df = pd.DataFrame(data)

X = df[['Area']]
y = df['Price']

model = LinearRegression()
model.fit(X, y)

area_values = np.array([5000, 8000, 9000]).reshape(-1, 1)
predictions = model.predict(area_values)

for area, price in zip(area_values.flatten(), predictions):
    print(f"Predicted price of a home with area = {area} sqr ft is ${price:,.2f}")

plt.scatter(df['Area'], df['Price'], color='green', label='Data points')
plt.plot(df['Area'], model.predict(X), color='black', label='Regression line')
plt.xlabel('Area (sqr ft)')
plt.ylabel('Price ($)')
plt.title('Linear Regression: Price vs Area')
plt.legend()
plt.show()
```
