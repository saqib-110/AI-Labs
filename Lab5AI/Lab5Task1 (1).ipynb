```python
import pandas as pd
import numpy as np
from sklearn.linear_model import LinearRegression
import matplotlib.pyplot as plt

data = {
    'area': [2600, 3000, 3200, 3600, 4000],
    'price': [550000,565000,610000,680000,725000]
}

df = pd.DataFrame(data)
print("Dataset Preview:")
print(df)

plt.xlabel('Area (sq ft)')
plt.ylabel('Price ($)')
plt.scatter(df['area'], df['price'], color='green', marker='o', label='Data Points')
plt.legend()
plt.show()

X = df[['area']]
y = df['price']

model = LinearRegression()
model.fit(X, y)

df['predicted_price'] = model.predict(X)

print("\nActual and Predicted Prices:")
print(df)

plt.scatter(df['area'], df['price'], color='green', marker='o', label='Actual Data')
plt.plot(df['area'], df['predicted_price'], color='black', linewidth=2, label='Regression Line')
plt.xlabel('Area (sq ft)')
plt.ylabel('Price ($)')
plt.legend()
plt.show()

print(f"\nSlope (m): {model.coef_[0]:.2f}")
print(f"Intercept (b): {model.intercept_:.2f}")
```
