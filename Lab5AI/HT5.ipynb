```python
import pandas as pd
import numpy as np
from sklearn.linear_model import LinearRegression
import matplotlib.pyplot as plt

# Updated dataset with 'area' and 'House_no' columns
data = {
    'area': [1800, 2200, 2700, 3100, 3500, 3900, 4200, 4700],
    'House_no': [3, 3, 4, 4, 5, 5, 5, 6],
}

df = pd.DataFrame(data)
print("Dataset Preview:")
print(df)

# Plotting the actual data points for 'House_no' prediction
plt.xlabel('Area (sq ft)')
plt.ylabel('House No')
plt.scatter(df['area'], df['House_no'], color='green', marker='o', label='Data Points')
plt.legend()
plt.show()

# Define the input (X) and output (y) variables for 'House_no' prediction
X = df[['area']]  # Features (area)
y = df['House_no']  # Target (House_no)

# Train the Linear Regression model
model = LinearRegression()
model.fit(X, y)

# Predicting the 'House_no' based on 'area'
df['predicted_house_no'] = model.predict(X)

print("\nActual and Predicted House Numbers:")
print(df)

# Plotting the actual data points and regression line for 'House_no' prediction
plt.scatter(df['area'], df['House_no'], color='green', marker='o', label='Actual Data')
plt.plot(df['area'], df['predicted_house_no'], color='black', linewidth=2, label='Regression Line')
plt.xlabel('Area (sq ft)')
plt.ylabel('House No')
plt.legend()
plt.show()

# Displaying the slope and intercept of the regression line for 'House_no'
print(f"\nSlope (m): {model.coef_[0]:.2f}")
print(f"Intercept (b): {model.intercept_:.2f}")
```
